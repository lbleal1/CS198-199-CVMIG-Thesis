```python
import cv2
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D
%matplotlib inline
```

```python
# open images
from tqdm import tqdm
img_dir = "assets/in_use/contrast/contrast_15_def/test/"
R = []
G = []
B = []

for fname in tqdm(sorted(os.listdir(img_dir))):
    print(fname)
    if fname == "tile_202.png":
        img = cv2.imread(img_dir + fname)
        for i in range(256):
            R = R + list(img[i][:,0]-123.7)
            G = G + list(img[i][:,1]-116.5)
            B = B + list(img[i][:,2]-103.9)
    

```

```python
# plot in one go
fig, ax = plt.subplots(figsize=(10,8))
ax = plt.axes(projection="3d")
ax.set_xlim3d(-200, 300)
ax.set_ylim3d(-200,300)
ax.set_zlim3d(-200,300)
ax.scatter3D([0], [0], [0], color = "red")
ax.scatter3D(R, G, B, label='A')
ax.set(xlabel='Red', ylabel='Green', zlabel='Blue')
#ax.scatter3D(df['Se_X'], df['Se_Y'], df['Se_Z'], label='Se1');

#ax.legend()
plt.show()
```

```python
# get the mean of the whole dataset
import numpy as np
from tqdm import tqdm
import os

img_dir = "assets/in_use/contrast/contrast_def_15/"
R_test = 0
G_test = 0
B_test = 0
# test
for fname in tqdm(os.listdir(img_dir+"test/")):
    if fname[-4:] == ".png":
        img = cv2.imread(img_dir+"test/" + fname)
        R_test = R_test + np.sum(img[:,:,0])
        G_test = G_test + np.sum(img[:,:,1])
        B_test = B_test + np.sum(img[:,:,2])
        #print(np.sum(img[:,:,0]))

numte = 200*256*256
R_temean = R_test / numte
G_temean = G_test / numte
B_temean = B_test / numte

print(R_temean)
print(G_temean)
print(B_temean)
```
